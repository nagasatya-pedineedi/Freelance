### Never drop or remove the data!!!!

np.nan - np.nan function is used to introduce null values in the data

```python
tab = pd.DataFrame(
                    {'Column1':[1,2,3,4,np.nan,np.nan],
                    'Column2':[1,2,3,4,22,33]}
)
tab
```

```python
import seaborn as sns
sns.boxplot(tab['Column1'])
```

```python
import matplotlib.pyplot as plt
plt.boxplot(tab['Column1'])
plt.show()
```

```python
# Numerical columns: Mean, median, arbitary 
tab['Column1'] = tab['Column1'].fillna(tab['Column1'].mean())
# mode, arbitary
```

```python
tab
```

```python
tab.dropna(axis=0)
```
