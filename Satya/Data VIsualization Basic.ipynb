- What is a boxplot?
Boxplot is used to check the outliers
- What are outliers in the data?
Outliers are the data points which vary significantly from the normal data points. There are positive and negative outliers. The data points that above and below the boxplot upper and lower boudaries are referred as outliers

```python
iqr = np.quantile(df['MSSubClass'], 0.75) - np.quantile(df['MSSubClass'],0.25)
iqr
```

```python
upper_limit = np.quantile(df['MSSubClass'], (0.75)) + (1.5*iqr)
lower_limit = np.quantile(df['MSSubClass'], (0.25)) - (1.5*iqr)
print("upper limit: ", upper_limit)
print("lower limit:", lower_limit)
```

##### Methods of plotting a bar chart

1. Using a group by

```python
# Grouping of the data based on the categorical column that we're specifying
# Aggregating the numerical columns values
df.groupby(['Foundation']).mean()
```

```python
g = df.groupby(['Foundation']).mean()
ax = g['TotalBsmtSF'].plot(kind='bar')
# for container in ax.containers:
#     plt.bar_label(container)
```

```python
sns.countplot(df['Foundation'].value_counts())
plt.show()
```

```python
df.groupby(['HouseStyle']).mean()
```

```python
g1 = df.groupby(['HouseStyle']).mean()
g1['MasVnrArea'].plot(kind='barh')
plt.title('HouseStyle Bar Chart')
plt.xlabel('Average MasVnrArea')
plt.show()
```
